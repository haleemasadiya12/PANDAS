#PANDAS

Pandas is one of the tools in Machine Learning which is used for data cleaning and analysis. It has features which are used for exploring, cleaning, transforming and visualizing from data.

**What's pandas for?**

Through pandas, you get acquainted with your data by cleaning, transforming, and analyzing it.

*For example*, say you want to explore a dataset stored in a CSV on your computer. Pandas will extract the data from that CSV into a DataFrame.

**Advantages of Pandas Library**

1.Data representation

2.Less writing and more work done

3.Efficiently handles large data

4.Makes data flexible and customizable

**Disadvantages of Pandas Library**

1.Steep learning curve

2.Difficult syntax

3.Poor compatibility for 3D matrices

4.Bad documentation
